```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv", index_col = 0)
```

```python
# Start coding here! Use as many cells as you like
```

```python
netflix_df.head(5)
```

```python
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]
```

```python
movies_1990s = netflix_subset[(netflix_subset["release_year"] >= 1990) & 
                               (netflix_subset["release_year"] < 2000)]
```

```python
movies_1990s.head(5)
```

```python
plt.hist(movies_1990s["duration"], bins = 6)
plt.show()
```

```python
duration = 100
```

```python
action_movie = movies_1990s[movies_1990s["genre"] == "Action"]
```

# Starting counter using for loop

```python
short_movie_count = 0

for label, row in action_movie.iterrows():
    if row["duration"]<90:
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count
    print (short_movie_count)
```
